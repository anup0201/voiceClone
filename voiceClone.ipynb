## Import dependencies and initialise model

```python
import torch
import torchaudio
import torch.nn as nn
import torch.nn.functional as F

import IPython

from tortoise.api import TextToSpeech
from tortoise.utils.audio import load_audio, load_voice, load_voices

# This will download all the models used by Tortoise from the HF hub.
tts = TextToSpeech()
```

## Text to be spoken initialisation and laoding sample voice file

```python
text = "ChatGPT is smarter than a some of the lawyers based in US"
voice = 'anup'
voice_samples, conditioning_latents = load_voice(voice)
```

## Generate voice

```python
gen = tts.tts_with_preset(text, voice_samples=voice_samples, conditioning_latents=conditioning_latents, 
                          preset='fast')
```

```python
torchaudio.save(f'generated-anup.wav', gen.squeeze(0).cpu(), 24000)
IPython.display.Audio(f'generated-anup.wav')
```
